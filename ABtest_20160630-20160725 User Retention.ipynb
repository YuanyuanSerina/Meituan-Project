```python
def cal_visit_ratio(df,day_num):
    df_visit_curve = df.groupby(['t.pricetypeint']).agg(
                                            {
                                            'f0630': np.sum,
                                            'f0701': np.sum,
                                            'f0702': np.sum,
                                            'f0703': np.sum,
                                            'f0704': np.sum,
                                            'f0705': np.sum,
                                            'f0706': np.sum,
                                            'f0707': np.sum,
                                            'f0708': np.sum,
                                            'f0709': np.sum,
                                            'f0710': np.sum,
                                            'f0711': np.sum,
                                            'f0712': np.sum,
                                            'f0713': np.sum,
                                            'f0714': np.sum,
                                            'f0715': np.sum,
                                            'f0716': np.sum,
                                            'f0717': np.sum,
                                            'f0718': np.sum,
                                            'f0719': np.sum,
                                            'f0720': np.sum,
                                            'f0721': np.sum,
                                            'f0722': np.sum,
                                            'f0723': np.sum,
                                            'f0724': np.sum,
                                            'f0725': np.sum
                                            }).reset_index()
    df_visit_curve.rename(columns = {
                                            'f0630': '0630访问人数',
                                            'f0701': '0701访问人数',
                                            'f0702': '0702访问人数',
                                            'f0703': '0703访问人数',
                                            'f0704': '0704访问人数',
                                            'f0705': '0705访问人数',
                                            'f0706': '0706访问人数',
                                            'f0707': '0707访问人数',
                                            'f0708': '0708访问人数',
                                            'f0709': '0709访问人数',
                                            'f0710': '0710访问人数',
                                            'f0711': '0711访问人数',
                                            'f0712': '0712访问人数',
                                            'f0713': '0713访问人数',
                                            'f0714': '0714访问人数',
                                            'f0715': '0715访问人数',
                                            'f0716': '0716访问人数',
                                            'f0717': '0717访问人数',
                                            'f0718': '0718访问人数',
                                            'f0719': '0719访问人数',
                                            'f0720': '0720访问人数',
                                            'f0721': '0721访问人数',
                                            'f0722': '0722访问人数',
                                            'f0723': '0723访问人数',
                                            'f0724': '0724访问人数',
                                            'f0725': '0725访问人数'          
                                            }, inplace=True)
    for i in range(day_num+1,26):
        
        if(day_num==0):
            if(i<10):
                df_visit_curve['070'+str(i)+'访问人数占比'] = 100.0 *df_visit_curve['070'+str(i)+'访问人数'] / df_visit_curve['0630访问人数']
            else:
                df_visit_curve['07'+str(i)+'访问人数占比'] = 100.0 *df_visit_curve['07'+str(i)+'访问人数'] / df_visit_curve['0630访问人数']
        elif(day_num>=10):
                df_visit_curve['07'+str(i)+'访问人数占比'] = 100.0 *df_visit_curve['07'+str(i)+'访问人数'] / df_visit_curve['07'+str(day_num)+'访问人数']
        else:
            if(i<10):
                df_visit_curve['070'+str(i)+'访问人数占比'] = 100.0 *df_visit_curve['070'+str(i)+'访问人数'] / df_visit_curve['070'+str(day_num)+'访问人数']
            else:
                df_visit_curve['07'+str(i)+'访问人数占比'] = 100.0 *df_visit_curve['07'+str(i)+'访问人数'] / df_visit_curve['070'+str(day_num)+'访问人数']
        
        df_visit_curve_2 = df_visit_curve.iloc[:,27:].copy()     
        df_visit_curve_2T = df_visit_curve_2.T
        df_visit_curve_2T.rename(columns = {
                                0 : '固定',
                                1 : '阶梯',
                                2 : '动态'
                                 }, inplace=True)
    
    return df_visit_curve_2T
```
