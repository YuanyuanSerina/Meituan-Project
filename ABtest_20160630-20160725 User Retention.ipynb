```python
def cal_visit_ratio(df,day_num):
    df_visit_curve = df.groupby(['t.pricetypeint']).agg(
                                            {
                                            'f0630': np.sum,
                                            'f0701': np.sum,
                                            'f0702': np.sum,
                                            'f0703': np.sum,
                                            'f0704': np.sum,
                                            'f0705': np.sum,
                                            'f0706': np.sum,
                                            'f0707': np.sum,
                                            'f0708': np.sum,
                                            'f0709': np.sum,
                                            'f0710': np.sum,
                                            'f0711': np.sum,
                                            'f0712': np.sum,
                                            'f0713': np.sum,
                                            'f0714': np.sum,
                                            'f0715': np.sum,
                                            'f0716': np.sum,
                                            'f0717': np.sum,
                                            'f0718': np.sum,
                                            'f0719': np.sum,
                                            'f0720': np.sum,
                                            'f0721': np.sum,
                                            'f0722': np.sum,
                                            'f0723': np.sum,
                                            'f0724': np.sum,
                                            'f0725': np.sum
                                            }).reset_index()
    df_visit_curve.rename(columns = {
                                            'f0630': '0630',
                                            'f0701': '0701',
                                            'f0702': '0702',
                                            'f0703': '0703',
                                            'f0704': '0704',
                                            'f0705': '0705',
                                            'f0706': '0706',
                                            'f0707': '0707',
                                            'f0708': '0708',
                                            'f0709': '0709',
                                            'f0710': '0710',
                                            'f0711': '0711',
                                            'f0712': '0712',
                                            'f0713': '0713',
                                            'f0714': '0714',
                                            'f0715': '0715',
                                            'f0716': '0716',
                                            'f0717': '0717',
                                            'f0718': '0718',
                                            'f0719': '0719',
                                            'f0720': '0720',
                                            'f0721': '0721',
                                            'f0722': '0722',
                                            'f0723': '0723',
                                            'f0724': '0724',
                                            'f0725': '0725'          
                                            }, inplace=True)
    for i in range(day_num+1,26):
        
        if(day_num==0):
            if(i<10):
                df_visit_curve['070'+str(i)+''] = 100.0 *df_visit_curve['070'+str(i)+''] / df_visit_curve['0630']
            else:
                df_visit_curve['07'+str(i)+''] = 100.0 *df_visit_curve['07'+str(i)+''] / df_visit_curve['0630']
        elif(day_num>=10):
                df_visit_curve['07'+str(i)+''] = 100.0 *df_visit_curve['07'+str(i)+''] / df_visit_curve['07'+str(day_num)+'']
        else:
            if(i<10):
                df_visit_curve['070'+str(i)+''] = 100.0 *df_visit_curve['070'+str(i)+''] / df_visit_curve['070'+str(day_num)+'']
            else:
                df_visit_curve['07'+str(i)+''] = 100.0 *df_visit_curve['07'+str(i)+''] / df_visit_curve['070'+str(day_num)+'']
        
        df_visit_curve_2 = df_visit_curve.iloc[:,27:].copy()     
        df_visit_curve_2T = df_visit_curve_2.T
        df_visit_curve_2T.rename(columns = {
                                0 : '',
                                1 : '',
                                2 : ''
                                 }, inplace=True)
    
    return df_visit_curve_2T
```
